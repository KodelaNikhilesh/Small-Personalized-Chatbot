```python
import nltk
import random
from nltk.chat.util import Chat, reflections # Changed 'chat' to 'Chat'
```

```python
paris = [

    [r"my name is (.*)", ["Hello %1, how can I help you?"]],
    [r"hi|hello|hey", ["Hello!", "Hey there!", "Hi! How can I assist you?"]],
    [r"how are you", ["I'm just a bot, but I'm doing great! How about you?"]],
    [r"what is your name", ["I'm a chatbot, and I'm here to chat with you!"]],
    [r"where are you from", ["I'm from the digital world!"]],
    [r"what can you do", ["I can chat with you, answer simple questions, and make your day better!"]],
    [r"who created you", ["I was created by a developer who loves AI and Python!"]],
    [r"what is the weather like", ["I'm not connected to live weather data, but you can check a weather website!"]],
    [r"bye|goodbye", ["Goodbye! Have a great day!", "See you later!"]],
    [r"how old are you", ["I'm as old as the code that runs me!"]],
    [r"what is love", ["Love is a complex emotion that makes life beautiful!"]],
    [r"tell me a joke", ["Why don’t programmers like nature? It has too many bugs!"]],
    [r"who is your favorite person", ["I like everyone equally!"]],
    [r"can you help me", ["Of course! What do you need help with?"]],
    [r"what is (.*)\?", ["I'm not sure, but I can try to help!"]],
    [r"you have feelings (.*)", ["I'm sorry to say but,Not"]],
    [r"i am (.*) today", ["I hope you have a great day despite feeling %1!"]],
    [r"do you like music", ["I can't listen to music, but I know many people love it!"]],
    [r"what is your favorite movie", ["I don't watch movies, but I hear classics like Inception are great!"]],
    [r"what is python", ["Python is a powerful programming language known for its simplicity and versatility!"]],
    [r"what is java", ["Java is a widely-used programming language for coding web applications"]],
    [r"do you know AI", ["Yes! AI stands for Artificial Intelligence, and I'm an example of it."]],
    [r"(.*)", ["I'm not sure how to respond to that. Can you rephrase?"]]

]

chatbot = Chat(paris, reflections)

def main():
  print("Hi, I'm Chatbox. How can I assist you?")
  while True:
    user_input = input("You: ")
    response = chatbot.respond(user_input)
    print("Chatbot:", response)
    if user_input.lower() == 'quit':
      break
```

```python
main()
```
